```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  # numpy

# 
data = {
    'Model': ['Baseline BART', 'Baseline BART', 'Baseline BART', 'Baseline BART', 'Baseline BART', 'RMT-Summ', 'RMT-Summ', 'RMT-Summ', 'RMT-Summ', 'RMT-Summ'],
    'Seq_len': [512, 1024, 2048, 4096, 8192, 512, 1024, 2048, 4096, 8192],
    'Mem': [2679, 4195, 9182, 27200, 'OOM', 2685, 4072, 7051, 12997, 24881],
}

# DataFrame
df = pd.DataFrame(data)

# 
custom_palette = ["#cc0033", "#003366"]

# Seaborn
# Seabornwhitegrid
sns.set(style="white", palette=custom_palette)

# 
sns.set_style({"grid.linewidth": 0.5, "grid.alpha": 0.5, "axes.grid": True, "axes.grid.axis": "x"})

# 
sns.set_style({"axes.grid.which": "both", "axes.grid.axis": "y", "grid.linestyle": "-"})

# 'Mem''OOM'
df['Mem'].replace('OOM', np.nan, inplace=True)

# 
plt.figure(figsize=(10, 6))
bar_width = 0.3  # 
for i, (model, group_df) in enumerate(df.groupby('Model')):
    seq_len_range = [x + i*bar_width for x in range(1, len(group_df['Seq_len']) + 1)]
    plt.bar(seq_len_range, group_df['Mem'], width=bar_width, label=model, alpha=0.7)

# 
plt.title('Train GPU Memory Usage', fontsize=24)
plt.xlabel('Input Size')
plt.ylabel('MB')
plt.xticks([x + bar_width/2 for x in range(1, len(df['Seq_len'].unique()) + 1)], df['Seq_len'].unique())  # 
plt.legend()

# 
plt.show()
```
