```python
# 把所有的逗号和句号进行统一
def replace_comma(text):
    text = text.replace(' ', '　') # JUMAN++无法处理半角空格
    text = text.replace('@', '') # JUMAN++似乎存在bug，会错误读取@
    text = text.replace('#', '') # JUMAN++似乎存在bug，会错误读取#
    # text = text.replace(',', '，')
    # text = text.replace('.', '．')
    text = text.replace('\n', 'linefeed') #无法处理换行符
    text = text.replace('\u3000', '') 
    return text
```

```python
df[['text', 'summary', 'incremental_summary']] = \
    df[['text', 'summary', 'incremental_summary']].map(replace_comma)
```

```python
from pyknp import Juman
jumanpp = Juman()

# preprocess by JUMAN++
def preprocess(text):
    mlist = jumanpp.analysis(text)
    return ' '.join([mrph.midasi for mrph in mlist.mrph_list()])
```

```python
from pyknp import Juman
jumanpp = Juman()

# preprocess by JUMAN++
def preprocess(text):
    split = text.split('、') # 按照．进行句子分割
    output = []
    for num, seq in enumerate(split):
        mlist = jumanpp.analysis(seq)
        output.append(' '.join([mrph.midasi.strip() for mrph in mlist.mrph_list()]))
    return ' 、 '.join(output).strip()
```

```python
import pandas as pd

df = pd.read_json('./datasets/tobyoki/tobyoki-event_summary_juman_processed_grouped.json', encoding='utf-8')
```

```python
def replace_back(texts):
    output = []
    for text in texts:
        text = text.replace("linefeed", "\n")
        output.append(text)
    return output

df[['text', 'summary', 'incremental_summary']] = \
    df[['text', 'summary', 'incremental_summary']].map(replace_back)
```
